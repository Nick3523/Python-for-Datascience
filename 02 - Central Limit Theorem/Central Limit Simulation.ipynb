# Central Limit Theorem in Python


Le théorème central limite (aussi improprement appelé théorème de la limite centrale ou centrée) établit la convergence en loi de la **somme** d'une suite de variables aléatoires (indépendantes, et identiquements distribuées) vers **la loi normale**. 

Intuitivement, ce résultat affirme que toute somme de variables aléatoires indépendantes tend dans certains cas vers une variable aléatoire gaussienne.

## Graphique de la loi normal sur l'intervalle [-100,100]

```python
%matplotlib inline  
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt, pi, exp
import pylab 
```

```python
domaine = range(-100,100)
mu = 0
sigma = 20


f = lambda x : 1/(sqrt(2*pi*pow(sigma,2))) * exp(-pow((x-mu),2)/(2*pow(sigma,2)))

y = [f(x) for x in domaine]


plot = plt.plot(domaine, y)
```

## Simulation de variables aléatoires indépendantes, identiquement distribuées (iid)

```python
#Create an array of the given shape and populate it with random samples from a uniform distribution over [0, 1).
vecteur_aleatoire = np.random.rand(30) 
plot = plt.scatter(range(30),vecteur_aleatoire)
```

```python
matrice_aleatoire_a_imprimer = np.random.rand(3,5)
print(matrice_aleatoire_a_imprimer)
matrice_aleatoire_a_grapher = np.random.rand(2,30)
fig = plt.figure()
#Le signe ":" demande à Numpy de nous donner toutes les colonnes
plot1 = plt.scatter(range(30),matrice_aleatoire_a_grapher[0,:])
plot2 = plt.scatter(range(30),matrice_aleatoire_a_grapher[1,:])
```

```python
# Nous allons considérer 100 variables aléatoires, 
# chacune avec 200 échantillons.
matrice_aleatoire = np.random.rand(100,200) #une matrice de 100 lignes, 200 colonnes

sommes = np.sum(matrice_aleatoire, 0) #0 pour effectuer la somme en colonnes, 1 l'effectue en ligne, et rien effectue simpelement la somme
```

```python
plot = plt.scatter(range(200), sommes)
```

### Avec un histogramme,  et pour un échantillon de  100*200, on commence à voir une ressemblance avec la courbe en cloche :

```python
plot = plt.hist(sommes)
```

### Le théorème de central limite, fait tendre le nombre de réalisations à l'infini, ici avec 9000x9000, nous constatons déjà un rapprochement avec la courbe de la loi normale

```python
matrice_aleatoire = np.random.rand(9000,9000) #génération d'un échantillion sur une matrice 9000x9000 suivant une loi uniforme
sommes = np.sum(matrice_aleatoire,0) #On somme les colonnes
plot = plt.hist(sommes, bins=100)
```
